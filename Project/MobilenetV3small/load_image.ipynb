```python
import torch
from torchvision import transforms
```

```python
# Loading an Image
import cv2
import numpy as np

image1 = cv2.imread('Image_Samples/n01665541_leatherback_turtle.JPEG')
image2 = cv2.imread('Image_Samples/n01514668_cock.JPEG')
image3 = cv2.imread('Image_Samples/n01440764_tench.JPEG')


#Plot the images
import matplotlib.pyplot as plt

plt.imshow(image1)
plt.show()

plt.imshow(image2)
plt.show()

plt.imshow(image3)
plt.show()
```

# Image 1

```python
# image1 = torch.tensor(image1)
# print(image1.shape)

transform = transforms.Compose([transforms.ToTensor(), transforms.Resize((224, 224))])

image_1 = transform(image1)

## Saving the images as flattened lists in .h files

with open("test_data_class34.h","w") as file:
    file.write("""#include <stdint.h>

using namespace std;
               
static const double test_image_class34[224*224*3] = {""")
    
    image = image_1.permute(1, 2, 0).flatten().numpy()


    for i in range(len(image)):
        file.write(str(image[i]) + ", ")
    file.write("};")
```
